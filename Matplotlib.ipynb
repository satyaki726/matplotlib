```python
def comparator(batsman1,batsman2):
    big=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=big[big['batsman']==batsman1]
    batsman_data1=big[big['batsman']==batsman2]
    p=batsman_data.groupby('season').sum()['batsman_runs']
    p1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(p.index, p.values,color="green",marker="d",label=batsman1)
    plt.plot(p1.index, p1.values,color="yellow",marker="+",label=batsman2)
    plt.xlabel('Season')
    plt.ylabel('Runs Scored')
    plt.legend()
```

```python
comparator("V Kohli","MS Dhoni")
```
